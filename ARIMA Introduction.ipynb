### ARIMA (auot regressive intergrated moving average)
- **it is generalized model from ARMA**
 - It actually devided into 3 compenents as **AR , I, MA**
- **Note** : ARIMA models work well on only time series data where data directly in relation with only time but in general if we consider stock price and other real time data then this model won't give better results because that data depends on many other outside factors (input attributes) 
- In **ARIMA**:
 - 1. **non seasonal ARIMA**
 - 2. **seasonal ARIMA (SARIMA)**
 - 3. **SRIMAX** (X means if we add input attributes)
- **ARIMA models can be applied on non-stationary data where we have to apply differentiation to get it into stationary**
- ARIMA models generally denoted with **ARIMA(p,d,q) i.e (p,d,q) are non negative integers**
- AR - p
- I  - d - diffrentiating component in order to get the series into stationary 
- MA - q
- Before applying ARIMA models we have to understand stationary and non-stationary
- **stationary** : constant mean and variance over time
- **non-stationary** : if not a constant mean and variance over time
